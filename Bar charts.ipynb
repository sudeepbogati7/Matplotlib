# Bar charts 

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np 
```

```python
labels = ['Nepal', 'India','Japan', 'China']
values = [20,50,30,90]

plt.title("Bar Graph", fontdict = {'fontname': 'Comic Sans MS','fontsize': 20})


plt.xlabel('lables')
plt.ylabel('values')
bars = plt.bar(labels, values)

#set hatch values // customize the each bar

# bars[0].set_hatch('/')
# bars[1].set_hatch('|')
# bars[2].set_hatch('-')
# bars[3].set_hatch('*')

# alternatively: 
patterns = ['-' , '/' , '|' , '.']
for bar in bars:
    bar.set_hatch(patterns.pop(0))

    
#resize the graph : 
#plt.figure(figsize = (6,4))

#show the graph only 
plt.show()
```
