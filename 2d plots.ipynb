```python
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
```

# 2D plots 

```python
fig = plt.figure()
x = np.array([1,2,3,4,5])

fig, axes = plt.subplots(1,4, figsize = (16,5))\


axes[0].set_title('scatter plot')
axes[0].scatter(x, x+ 0.25*np.random.randn(len(x)), color = 'r')


axes[1].set_title('Step plot')
axes[1].step(x,x+3,lw = 2, color = 'g', marker = '.')


axes[2].set_title('bar plot')
axes[2].bar(x, x**2,width = 0.5,align = 'center' )


axes[3].set_title('fill between ')
axes[3].fill_between(x, x**2,x**3, color = 'y', alpha = 0.8)



```

## Radar plot

```python
# ----------------------- radar chart -------------------------


fig = plt.figure(figsize = (7,7))

ax = fig.add_axes([0,0,0.7,0.7], polar = True)
t = np.linspace(0,2*np.pi, 100)
ax.plot(t,t**0.5, color = 'red', lw = 2)

plt.show()
```

### Histograms

```python
# ------------- Histogram ---------------------------

n = np.random.randn(100000)
fig, axes = plt.subplots(1,2, figsize = (12,5))

axes[0].set_title('Default Histogram')
axes[0].hist(n)

axes[1].set_title('Cumulative detailed histogram ')
axes[1].hist(n, cumulative = True, bins = 50, color = 'g')
plt.show()
```
