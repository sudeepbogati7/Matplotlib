```python
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
```

```python
x = np.linspace(1,10,25)
y = np.sin(x)
z = np.cos(x)
w = np.tan(x)

fig = plt.figure()

sin = fig.add_axes([0,0.1,0.8,0.8])
cos = fig.add_axes([0.9,0.5,0.3,0.3])
tan = fig.add_axes([0.9,0.1,0.3,0.3])

sin.plot(x,y,'r.-.', label = 'sin')
cos.plot(x,z,'b.-', label = 'cos')
tan.plot(x,w,'g.-', label= 'tan')

fig.legend(loc =1)
plt.show()
```

```python
d = np.array([1,2,3,4,5,6,7])
fig = plt.figure()
fig, ax = plt.subplots()


line1, = ax.plot(d,d+2,'r', linestyle = '-.', label = 'd+2', alpha = 0.5, )
line1.set_dashes([1,3,5,6])

ax.plot(d,d+3,'b', label = 'd+3')
ax.plot(d,d+5,'g', label = 'd-2 & d+2')

ax.legend() 
plt.show()
```
