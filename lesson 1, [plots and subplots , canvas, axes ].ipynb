```python
n = np.arange(0,90,10)
x = np.sin(n)
y = np.cos(n)

plt.subplot(1,2,1)
plt.plot(x,n,'b')
plt.title('sin curve')
plt.xlabel('value')
plt.ylabel('degree')
plt.yticks(n[::30])


plt.subplot(1,2,2)
plt.plot(y,n,'r')
plt.title('cos curve')
```

# axis plot (canvas)

```python
m = [1,2,3,4]
n = [1,4,9,16]

fig = plt.figure()
axis1 = fig.add_axes([0.4,0.3,0.9,0.9])
axis2 = fig.add_axes([0.01,0.9,0.25,0.25])


axis1.plot(m,n,'b.-', label ='x^2')
axis1.set_xlabel('value of x')
axis1.set_ylabel('value of x²')
axis1.set_title('squared numbers of x ')
axis1.grid(True)
axis1.set_xlim(0,4.5)
axis1.set_ylim(0,20)

axis2.plot(m,n,'r*-',)



axis1.legend()
```

```python
fig, axes = plt.subplots(nrows = 1, ncols = 3)
for ax in axes:
    ax.plot(m,n,'b')
```

```python
# let's define another canvas 

fig = plt.figure()

axis1 = fig.add_axes([0.1,0.6,0.8,0.8])
axis2 = fig.add_axes([0.2,1,0.2,0.2])

axis1.plot(m,n,'g')
axis1.set_title('squares of the numbers ')


axis2.plot(n,m,'b')
axis2.set_title('square roots')
```

```python
# another alternative way 

fig = plt.figure(figsize = (10,6))


years = np.arange(2000,2023,3)

gold_rate = np.array([2,3,6,5,9,15,13,18,20,29,27,33])
silver_rate = np.array([1,3,2,6,4,2,5,8,9,8,13,12])

gold = fig.add_axes([0.0,0.6,0.5,0.5])
silver = fig.add_axes([0.7,0.6,0.5,0.5])


gold.plot(years, gold_rate,'y.-')
silver.plot(years, silver_rate,'#b4b8b5', marker= '.', label='silver')

gold.set_title('Price change in gold ')
gold.set_xlabel('year')
gold.set_ylabel('change rate')


#gold.set_xlim(years[0],years[11])



silver.set_title('Price change in silver ')
silver.set_xlabel('year')
silver.set_ylabel('change rate ')

plt.legend()
plt.show()
```
