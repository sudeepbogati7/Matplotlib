```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = [1,2,3,4,5]
y = [2,4,6,8,10]

# resize your graph
plt.figure(figsize = (5,3), dpi = 100) #dpi specifies pixels per inches , it can be changed separaretely while saving the figure


#keyboard argument notations : 
#plt.plot(x,y, label = '2x', color = "blue", linewidth = 1, linestyle = ':', marker = '.' , markersize = 10, markeredgecolor = 'red')  # creates a plot or graph , customize the graph------ linewidth = width of graph lne -----  marker = dot point on the graph    

# shorthand notation: 
# plt.plot('[color],[marker],[line]')
plt.plot(x,y,'ro--',linewidth = .5,label ='2x')


# Second line 

# select the intervals you want to plot points at : 
x2 = np.arange(0,5,0.5)

plt.plot(x2[:7], x2[:7]**2,'b', label = "x^2")
plt.plot(x2[6:],x2[6:]**2,'y--')

# grraph title and its customizations : 
plt.title("My First Graph", fontdict ={'fontname': 'Comic Sans MS', 'fontsize': 20})    # graph title 


#label the axises: 
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

#X, Y tickmarks (scale of the graph ):
#plt.xticks([1,2,3,4,5])   
#plt.yticks([3,6,9,12,15])


#add a legend  
plt.legend()    # connected with the -> label = "2x"........ <- in the plt.plot() function


#--------save your graph --------
plt.savefig("myFirstGraph.png", dpi =300) # here you can save your figure with separate dpi than live visualization 


#show the plot 
plt.show()
```
