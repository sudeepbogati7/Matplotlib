#### * Write a python program to create a pie chart of the popularity of programming languages 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#---------------data to plot -----------
languages = ['python', 'Java', 'C', 'JavaScript', 'R','PHP','c++']
popularity = [40,20,25,23,5,20,15]
```

```python
#explode the 1st slice 

explodes = (0.2,0,0,0,0,0,0)

plt.style.use('ggplot')
plt.pie(popularity, explode = explode, shadow = True, startangle = 200, labels = languages, autopct = '%.1f%%', pctdistance = 0.6)

plt.show()
```
