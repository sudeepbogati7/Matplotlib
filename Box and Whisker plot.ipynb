```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

##### Using the FIFA 2018 data (source : Kaggle)


```python
fifa = pd.read_csv('fifa_data.csv')
barcelona = fifa.loc[fifa.Club == 'FC Barcelona']['Overall']
real_madrid = fifa.loc[fifa.Club == 'Real Madrid']['Overall']
juventus = fifa.loc[fifa.Club == 'Juventus']['Overall']

plt.figure(figsize = (5,7))
plt.title('Professional soccer Team Comparison')
plt.ylabel('FIFA overall ratings ')

labels = ['FC Barcelona', 'Real Madrid', 'Juventus']
boxes = plt.boxplot([barcelona, real_madrid, juventus], labels = labels, patch_artist = True, medianprops = {'linewidth' : 1.5, 'color':'r'})   #patch_artist gives permission for facecolor
#medianprops is for the line inside the box 
for box in boxes['boxes']:
    #set edge color:
    box.set(color = '#3f93c4', linewidth = 1.5)
    
    #change fill color :
    box.set(facecolor = '#d5e0e6')
plt.style.use('default')

plt.show()
```
