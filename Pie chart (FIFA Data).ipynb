```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

### Pie Chart #1

```python
fifa = pd.read_csv('fifa_data.csv')
fifa
```

```python
left = fifa.loc[fifa['Preferred Foot'] == 'Left'].count()[0]
right = fifa.loc[fifa['Preferred Foot'] == 'Right'].count()[0]
```

```python
labels = ['left', 'right']
colors = ['#76cca3', '#b54e56']


plt.pie([left, right], labels = labels , colors = colors, autopct = '%.3f %%')


plt.title('Foot preferece of FIFA players')
plt.show()
```

## Pie chart #2

```python
fifa.Weight = [ int (x.strip('lbs')) if type(x) == str else x for x in fifa.Weight ]
fifa.Weight[0]
```

```python
#division of data 

light  = fifa.loc[fifa.Weight < 125 ].count()[0]
light_medium = fifa.loc[(fifa.Weight >= 125) & (fifa.Weight < 150)].count()[0]
medium = fifa.loc[(fifa.Weight >= 150) & (fifa.Weight < 175)].count()[0]
medium_heavy = fifa.loc[(fifa.Weight >= 175) & (fifa.Weight < 200)].count()[0]
heavy = fifa.loc[fifa.Weight >= 200].count()[0]


weights = [light, light_medium , medium , medium_heavy, heavy]

labels = ['less than 125', '125-150', '150 - 175','175 - 200', 'greater than 200']


explode = [.4, .2 ,0, 0, .4]

plt.pie(weights,labels = labels, autopct = '%.1f%%', pctdistance = 0.7, explode = explode)
plt.style.use('ggplot')



plt.title('Weight Distribution of FIFA players (in lbs)', fontdict = {'fontweight': 'bold', 'fontfamily': 'Consolas', 'fontsize': 16})
plt.show()
```
